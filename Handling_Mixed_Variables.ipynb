```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('passenger trawling with')
```

```python
#Extract numerical data
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')

```

```python
#extract numerical data
df['number_catogorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)')
df['cabin_cat']=df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot()
```

```python

#extract the lastbit of ticket as num
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

#extract the first part of ticket as catergry
df['ticket_cat']=df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),df['cabin_cat'],df['ticket_cat'])
```
